```python
def term_finder(line):
    #returns the meanings of words in a line of text
    terms = re.findall(r"(?<=\[)(.*?)(?=\])", line)
    terms += re.findall(r"(?<=\:)([^:]*?)(?=\[1])", line)
    terms = [term for term in terms if term != '1']
    return terms

def list_flattening(pn_list):
    return [noun for nouns in pn_list for noun in nouns]

def text_splitter(table):
    col = 'text'
    if col not in table.labels:
        col = ' ' + col
    translation = table.apply(term_finder, col)
    return list_flattening(translation)
    
# def text_distances(text_list):
#     from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
#     from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
#     from sklearn.metrics import pairwise
    
#     cv = CountVectorizer(stop_words = ENGLISH_STOP_WORDS)
#     dtm = cv.fit_transform(text_list)
#     tt = TfidfTransformer(norm='l1',use_idf=False)
#     dtm_tf = tt.fit_transform(dtm)

#     word_list = cv.get_feature_names()
#     dtm_array = dtm_tf.toarray()
#     return Table(word_list).with_rows(dtm_array)
```

```python
corpus_path = "corpus/"
def text_measures_alt(text_name):
    text = Table.read_table(corpus_path+text_name, ',')
    text_list = text_splitter(text)
    return text_list

def contains_digit(w):
    return np.any([char.isdigit() or char == '.' for char in w])
```

```python
with open('text_list.txt', 'r') as file_in:
        corpus = file_in.read().split()

corpus_table = Table(['Title', 'Text List'])
rows = []
for text_name in corpus:
    text_list = text_measures_alt(text_name)
    rows.append([text_name, text_list])

corpus_table = corpus_table.with_rows(rows)
vocabulary = list_flattening(corpus_table['Text List']) 
vocabulary = list(set([word for word in vocabulary if contains_digit(word) != True]))
corpus_table
```

```python
# Calculate distances among texts in vector space
from sklearn.metrics import pairwise

def word_counts(vocab, book):
    row_count = []
    for word in vocab:
        row_count.append(book.count(word))
    return row_count

dtm = []

for book in corpus_table['Text List']:
    dtm.append(word_counts(vocabulary, book))
    
docu_term_matrix = Table(vocabulary).with_rows(dtm)
docu_term_matrix = docu_term_matrix.with_column('book', list(corpus_table['Title'])).move_to_start('book')
docu_term_matrix
# dist_matrix = pairwise.pairwise_distances(dtm_tf, metric='euclidean')

# title_list = ['Book '+str(i+1) for i in range(len(one_list))]
# Table(title_list).with_rows(dist_matrix)
```

# Normalization

```python
# In order to make apples-to-apples comparisons across Books, we can normalize our values
# by dividing each word count by the total number of words in its Book.

desparse = docu_term_matrix.drop('book').to_df()
row_sums = np.array(np.sum(desparse, axis=1).tolist())
rows = docu_term_matrix.drop('book').to_array()
for i in range(len(rows)):
    rows[i] = list(rows[i])
    rows[i] = rows[i]/row_sums[i]
```
